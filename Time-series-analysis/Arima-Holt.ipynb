#### Parse Date

```r
setAs("character","myDate", function(from) as.Date(as.yearmon(from)) )
```

```r
data <- read.csv(file="./TimeSeriesAnalysisWithPython-master/data/AirPassengers.csv", colClasses=c('myDate', 'numeric'), header=T)
```

#### Holt Winters

```r
ts <- ts(data = data$NoPassengers, frequency=12)
```

```r
WTI.hwm <- HoltWinters( ts )
```

```r
#plot(WTI.hwm)
```

```r
WTI.hwf <- forecast:::forecast.HoltWinters( WTI.hwm, h=50 )
```

```r
summary( WTI.hwf )
```

```r
forecast:::plot.forecast( WTI.hwf )
```

#### ARIMA

```r
WTI.arima <- arima( data$NoPassengers, order=c(4,1,2), seasonal = list(order = c(4,1,2), period = 12),method="ML" )
```

```r
WTI.arimaFore <- forecast:::forecast.Arima( WTI.arima, h=50)
```

```r
forecast:::plot.forecast( WTI.arimaFore )
```
